```python
import pandas as pd
df = pd.read_csv("sales_data_with_discounts.csv")
df
```

```python
df.dtypes
```

```python
print(df.shape),print(df.columns)
```

```python
#####  Numarical columns
numerical_cols = df.select_dtypes(include=['number'])
print("Numerical columns:")
print(numerical_cols.columns)
```

```python
###Mean Value
mean_values = numerical_cols.mean()
mean_values
```

```python
####Median value
median_values = numerical_cols.median()
median_values
```

```python
###Mode Value 
mode_values = numerical_cols.mode().iloc[0]  
mode_values
```

```python
###Standard deviation 
std_values = numerical_cols.std()
std_values
```

```python
### Data Frame
statistics_df = pd.DataFrame({'Mean':mean_values,
                              'Median':median_values,
                              'Mode':mode_values,
                              'Standard Deviation':std_values
                             })
print(statistics_df)
```

```python
####Data Visualization####
###Histograms Plot
```

```python
import matplotlib.pyplot as plt
numerical_cols.hist(figsize=(14, 8), bins=10, edgecolor='black')
plt.suptitle("Histograms for Numerical Columns", fontsize=16)
plt.tight_layout()
plt.show 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Select numerical columns
numerical_cols = df.select_dtypes(include='number')
for col in numerical_cols.columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], bins=10, kde=True, color='skyblue', edgecolor='black')
# Calculate skewness
    skewness = df[col].skew()
    plt.title(f'Histogram of {col} (Skewness = {skewness:.2f})')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.axvline(df[col].mean(), color='red', linestyle='--', label='Mean')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
#### Box Plot of Numarical columns
import seaborn as sns
for col in numerical_cols.columns:
    plt.figure(figsize=(6, 1.5))
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot for {col}')
    plt.xlabel(col)
    plt.grid(True, axis='x')
    plt.tight_layout()
    plt.show()
```

```python
###Categorical columns
categorical_cols = df.select_dtypes(include='object').columns
print("Categorical columns:", list(categorical_cols))
```

```python
###Categorical columns of Categorical columns
import seaborn as sns
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot for {col}')
    plt.xlabel(col)
    plt.grid(True, axis='x')
    plt.tight_layout()
    plt.show()
```

```python
### Bar Chart Analysis for Categorical Column 
import matplotlib.pyplot as plt
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    df[col].value_counts().plot(kind='bar', color='orange', edgecolor='black')
    plt.title(f'Bar Chart - {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
####Standardization of Numerical Variables
```

```python
from sklearn.preprocessing import StandardScaler
numerical_cols.hist(figsize=(12, 10), edgecolor='black',color='red')
plt.suptitle("Before Standardization")
plt.tight_layout()
plt.show()
####Standardize using z-score
scaler = StandardScaler()
standardized_array = scaler.fit_transform(numerical_cols)
df_standardized = pd.DataFrame(standardized_array, columns=numerical_cols.columns)
df_standardized.hist(figsize=(12, 10),edgecolor='black', color='lightgreen')
plt.suptitle("After Standardization")
plt.tight_layout()
plt.show()
```

```python
###Conversion of Categorical Data into Dummy Variables
categorical_col= pd.get_dummies(df,columns=categorical_cols,drop_first=False)
print(categorical_col.head())
```
