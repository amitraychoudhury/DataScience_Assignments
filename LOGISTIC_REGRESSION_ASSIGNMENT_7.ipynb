```python
df_train_non_null = df_train.fillna(" ")
df_train_non_null
```

```python
# Convert categorical column 'Sex' to numerical. Using Label Encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_train['Sex'] = le.fit_transform(df_train['Sex'])
df_train['Sex']
```

```python
# Dropping unwanted columns (feature extraction)
x_train = df_train.drop(['PassengerId','Name','Ticket','Cabin','Embarked','Survived'],axis=1)
x_train
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
df_imputer = pd.DataFrame(imputer.fit_transform(x_train),columns=x_train.columns)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(df_imputer)
```

```python
y_train = df_train.Survived
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x_train_scaled,y_train,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model_test = LogisticRegression()
model_test.fit(xtest,ytest)
```

```python
model_test_ypred=model_test.predict(xtest)
```

```python
from sklearn.metrics import classification_report
model_test_cr = classification_report(ytest,model_test_ypred)
print(model_test_cr)
```

```python
print(f"F1 Score for testing dataset is {0.85*100 : .2f}")
print(f"Accuracy for testing dataset is {0.85*100 : .2f}")
```

```python
from sklearn.metrics import classification_report
cr = classification_report(ytest,model_test_ypred)
print(cr)
```

```python
#2. What is cross-validation, and why is it important in binary classification?
#Cross-Validation:
#A technique where the dataset is split into k parts.
#The model is trained on k-1 folds and tested on the remaining fold.
#This process is repeated k times, each fold serving once as the test set.
#The final performance is the average of all runs.
#Why important in binary classification?
#Prevents overfitting → ensures the model is not just memorizing training data.
#Reliable evaluation → performance is measured on multiple subsets, not just one train-test split.
#Better generalization → helps check how well the model will perform on unseen data.
```
